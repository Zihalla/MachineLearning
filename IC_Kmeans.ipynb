```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
```

#1º questão: Usando K = 2

```python
#  não é necessário normalizar porque as casas estão próximas 
X = np.array([[-3, 4], [2, 0], [4, -4], [4, 3], [0, 2], [2, 1], [1, -1], [-5, 3]])
kmeans = KMeans(n_clusters = 2, init = 'k-means++', n_init = 10, max_iter = 300) 
y_pred = kmeans.fit_predict(X)
print('centroid pertencente:', y_pred) #retorna a qual centroid pertence
print('média dos clusters:', silhouette_score(X, kmeans.labels_)) #define a média dos clusters
print('pontos dos centroids:', kmeans.cluster_centers_) # pontos dos centroids
```

```python
plt.scatter(X[:,1], X[:,0], c = y_pred, cmap='RdYlBu') #posicionamento dos eixos x e y
plt.xlim(-10, 7) 
plt.ylim(-10, 7) 
plt.grid(color='blue', linestyle='--', linewidth=0.2 ) 
plt.scatter(kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,0], s = 90, c = 'yellow') #posição de cada centroide no gráfico
plt.show()
```

##Elbow para avaliar qual o k mais adequado

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.datasets import make_blobs
fig, ax = plt.subplots(1, figsize=(9,5))

visualizer = KElbowVisualizer(kmeans, k=(2, 3, 4, 5, 6, 7, 8))
visualizer.fit(X)
```
